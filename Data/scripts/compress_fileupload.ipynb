```python
import pandas as pd

df = pd.read_parquet(r'../gefs/klga.parquet')
```

```python
keep = [
    'init_time',
    'ensemble_member',
    'lead_time',
    'temperature_2m',
    'maximum_temperature_2m',
    'minimum_temperature_2m',
    'relative_humidity_2m',
    'total_cloud_cover_atmosphere',
    'downward_short_wave_radiation_flux_surface',
    'downward_long_wave_radiation_flux_surface',
    'wind_u_10m',
    'wind_v_10m',
    'wind_u_100m',
    'wind_v_100m',
    'pressure_reduced_to_mean_sea_level',
    'pressure_surface',
    'geopotential_height_500hpa',
    'precipitable_water_atmosphere',
    'valid_time'
]

df_clean = df.copy()[keep]
df_clean.init_time = pd.to_datetime(df_clean.init_time)
df_clean.valid_time = pd.to_datetime(df_clean.valid_time)
df_clean.lead_time = pd.to_timedelta(df_clean.lead_time)
float_cols = df_clean.select_dtypes('float64').columns.tolist()
df_clean[float_cols] = df_clean[float_cols].astype('float32')


df_clean.to_parquet(
    "klga_compressed.parquet",
    compression="zstd",
    compression_level=9,
    index=False
)
```
